```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pylab as plt
```

```python
analysis_folder_name = os.listdir('sixth/')
col = ['steps', 'D0A1', 'D0A2', 'D0A3',
                'D1A1', 'D1A2', 'D1A3',
                'D2A1', 'D2A2', 'D2A3']
```

```python
df = pd.DataFrame(columns=['steps', 'D0A1', 'D0A2', 'D0A3',
                                    'D1A1', 'D1A2', 'D1A3',
                                    'D2A1', 'D2A2', 'D2A3'])

all_intensities = []
for i, val in enumerate(analysis_folder_name):
    step_name = int(re.findall(r'\d+', val)[0])
    file = open('fourth/{}'.format(val))
    lines = file.readlines()
    intensities = []
    for ii in range(len(lines)):
        nm_lst = lines[ii].split('[')[-1].split(']')[0].split(',')
        intensities.append([float(i) for i in nm_lst])

    intensities = [item for sublist in intensities for item in sublist]
    df.loc[i] = [step_name] + intensities
    
df['steps']=df['steps'].astype(int)
df.sort_values(by='steps', ascending=True, ignore_index=True)
```

```python
for ii in df.keys()[1:]:
    plt.plot(df[ii])
```
